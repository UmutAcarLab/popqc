```python
import re

names = [
    "adder_8",
    "barenco_tof_5",
    "gf2^128_mult",
    "gf2^32_mult",
    "gf2^6_mult",
    "mod5_4",
    "qcla_adder_10",
    "tof_10",
    "vbe_adder_3",
    "barenco_tof_10",
    "csla_mux_3",
    "gf2^131_mult",
    "gf2^4_mult",
    "gf2^7_mult",
    "mod_adder_1024",
    "qcla_com_7",
    "tof_3",
    "barenco_tof_3",
    "csum_mux_9",
    "gf2^163_mult",
    "gf2^5_mult",
    "gf2^8_mult",
    "mod_mult_55",
    "qcla_mod_7",
    "tof_4",
    "barenco_tof_4",
    "gf2^10_mult",
    "gf2^16_mult",
    "gf2^64_mult",
    "gf2^9_mult",
    "mod_red_21",
    "rc_adder_6",
    "tof_5"
]


def decompose_ccz(line):
    pattern = r"ccz\s+q\[\s*(\d+)\s*\],\s*q\[\s*(\d+)\s*\],\s*q\[\s*(\d+)\s*\];"
    match = re.search(pattern, line)
    q1 = match.group(1)
    q2 = match.group(2)
    q3 = match.group(3)
    result = []
    result.append(f"cx q[{q2}], q[{q3}];")
    result.append(f"tdg q[{q3}];")
    result.append(f"cx q[{q1}], q[{q3}];")
    result.append(f"t q[{q3}];")
    result.append(f"cx q[{q2}], q[{q3}];")
    result.append(f"tdg q[{q3}];")
    result.append(f"cx q[{q1}], q[{q3}];")
    result.append(f"t q[{q3}];")
    result.append(f"t q[{q2}];")
    result.append(f"cx q[{q1}], q[{q2}];")
    result.append(f"t q[{q1}];")
    result.append(f"tdg q[{q2}];")
    result.append(f"cx q[{q1}], q[{q2}];")
    return result

for name in names:
    with open(f"{name}.qasm") as f:
        lines = f.readlines()
    results = []
    for line in lines:
        line = line.strip()
        if "ccz" in line:
            results.extend(decompose_ccz(line))
        else:
            results.append(line)
    with open(f"{name}_decomposed.qasm", "w") as f:
        for line in results:
            f.write(line + "\n")
```
