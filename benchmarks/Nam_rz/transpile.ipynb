```python
import re
file_names=["8","16","32","64","128","256","512","1024","2048"]
def quipper_to_qasm(quipper_lines):
    regex_h="""QGate\["H"\]\((\d+)\) with nocontrol"""
    regex_cnot="""QGate\["not"\]\((\d+)\) with controls=\[\+(\d+)\] with nocontrol"""
    regex_rz="""QRot\["exp\(-i%Z\)",(-?\d+\.\d+e[+-]?\d+)\]\((\d+)\)"""

    quipper_lines = quipper_lines.strip().split("\n")
    n_qubits=quipper_lines[0].count("Qbit")

    qasm_lines = ["OPENQASM 2.0;", "include \"qelib1.inc\";", f"qreg q[{n_qubits}];"]
    quipper_lines=quipper_lines[1:]
    for line in quipper_lines:
        if re.match(regex_h, line):
            qasm_lines.append("h q[{}];".format(re.match(regex_h, line).group(1)))
        elif re.match(regex_cnot, line):
            qasm_lines.append("cx q[{}], q[{}];".format(re.match(regex_cnot, line).group(2), re.match(regex_cnot, line).group(1)))
        elif re.match(regex_rz, line):
            qasm_lines.append("rz({}) q[{}];".format(re.match(regex_rz, line).group(1), re.match(regex_rz, line).group(2)))
        elif "Output" in line:
            continue
        else:
            assert False, "Unrecognized line: {}".format(line)
            

        
    return "\n".join(qasm_lines)
for qubit in file_names:
    filename="QFTAdd"+qubit+"_before"
    quipper_code = open(filename).read()
    qasm_code = quipper_to_qasm(quipper_code)
    with open("QFTAdd"+qubit+".qasm", "w") as f:
        f.write(qasm_code)
# quipper_code = open("quipper_code_2048.txt").read()

# qasm_code = quipper_to_qasm(quipper_code)
# print(qasm_code)
```
