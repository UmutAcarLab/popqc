```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def quad_func(x, a, b, c):
    return a * x**2 + b * x + c
# scale with 1023
gates = np.array([1, 2, 4,8,16,32,64])
time_soam = np.array([6.0479426,10.575893,21.99631,42.76964,83.08443,157.0595,322.5246])
time_voqc = np.array([0.6038451194763184,0.9326462745666504,1.6638848781585693,3.83449649810791,9.799204349517822,32.06591582298279,144.71882367134094])

params_soam, cov = curve_fit(quad_func, gates, time_soam)
params_voqc, cov = curve_fit(quad_func, gates, time_voqc)

print(params_soam)
print(params_voqc)
gates_fit=np.array([i for i in range(200)])
time_soam_fit = quad_func(gates_fit, *params_soam)
time_voqc_fit = quad_func(gates_fit, *params_voqc)
plt.plot(gates, time_soam, 'ro', label='SOAM')
plt.plot(gates_fit, time_soam_fit, 'r', label='SOAM Fit')
plt.plot(gates, time_voqc, 'bo', label='VOQC')
plt.plot(gates_fit, time_voqc_fit, 'b', label='VOQC Fit')
plt.legend()
plt.show()
```
